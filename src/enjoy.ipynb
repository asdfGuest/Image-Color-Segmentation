```python
import cv2
import numpy as np
from image_segment import segment_image
```

```python
IMAGE_PATH = '../car.png'
SAVE_PATH = '../assets/car.png'

img = cv2.imread(IMAGE_PATH)
img = (img / 255.0).astype(np.float32)

img, _ = segment_image(img, k=8, epochs=16, stride=2, tol=0.00001, verbose=False)

img = (img * 255.0).astype(np.uint8)
cv2.imwrite(SAVE_PATH, img)
```

```python
VIDEO_PATH = '../nature.mp4'
SAVE_PATH = '../assets/nature.mp4'

# setting
k = 16
epochs = 4
stride = 64
scale = 0.4


# video capture object
cap = cv2.VideoCapture(VIDEO_PATH)

# get video info
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH) * scale)
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT) * scale)
fps = int(cap.get(cv2.CAP_PROP_FPS))

# video write object
fourcc = cv2.VideoWriter_fourcc(*'AVC1')
out = cv2.VideoWriter(SAVE_PATH, fourcc, fps, (frame_width, frame_height))


cnt = 0

c = None
while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # preprocess
    frame = cv2.resize(frame, None, fx=scale, fy=scale)
    frame = frame.astype(float) / 255.0
    # color segmentation
    frame, c = segment_image(frame, k, epochs, stride, c, verbose=False)
    
    # write video
    frame = (frame * 255.0).astype(np.uint8)
    out.write(frame)
    
    print('time : %f'%(cnt / fps))
    cnt += 1


# release
cap.release()
out.release()
cv2.destroyAllWindows()
```
